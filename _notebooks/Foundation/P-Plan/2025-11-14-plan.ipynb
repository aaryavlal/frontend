    <span style="color: #94a3b8; font-weight: normal; font-size: 0.75em; margin-left: auto;">(Dhyan)</span>
  </h3>
  
  <p style="color: #e2e8f0; line-height: 1.8; font-size: 1.1em; margin: 15px 0;">
    <strong style="color: #38bdf8;">Sequential Computing:</strong> Instructions processed one at a time; total time = sum of all steps
  </p>
  <p style="color: #e2e8f0; line-height: 1.8; font-size: 1.1em; margin: 15px 0;">
    <strong style="color: #38bdf8;">Parallel Computing:</strong> Multiple processors on one computer run tasks simultaneously
  </p>
  <p style="color: #e2e8f0; line-height: 1.8; font-size: 1.1em; margin: 15px 0;">
    <strong style="color: #38bdf8;">Distributed Computing:</strong> Multiple separate devices/computers work together on the same problem
  </p>

<div style="background: linear-gradient(135deg, #1e293b 0%, #334155 100%); padding: 30px; border-radius: 12px; margin: 25px 0; border-left: 6px solid #22d3ee; box-shadow: 0 4px 15px rgba(0,0,0,0.2);">
  <h3 style="color: #22d3ee; margin-top: 0; font-size: 1.8em; display: flex; align-items: center; gap: 10px;">
    <span style="background: #22d3ee; color: #0f172a; padding: 5px 15px; border-radius: 8px; font-weight: bold;">2</span>
    Module 2: Why Parallel/Distributed Computing? 
    <span style="color: #94a3b8; font-weight: normal; font-size: 0.75em; margin-left: auto;">(Lucas)</span>
  </h3>
  
  <p style="color: #e2e8f0; line-height: 1.8; font-size: 1.1em; margin: 15px 0;">
    <strong style="color: #22d3ee;">‚ö†Ô∏è Hardware Limits:</strong> Single processors hit speed limits (overheating issues)
  </p>
  <p style="color: #e2e8f0; line-height: 1.8; font-size: 1.1em; margin: 15px 0;">
    <strong style="color: #22d3ee;">üí∞ Efficiency:</strong> Parallel/distributed solutions save time and money
  </p>
  <p style="color: #e2e8f0; line-height: 1.8; font-size: 1.1em; margin: 15px 0;">
    <strong style="color: #22d3ee;">üìà Scalability:</strong> Better scalability for handling increased workloads
  </p>
  <p style="color: #e2e8f0; line-height: 1.8; font-size: 1.1em; margin: 15px 0;">
    <strong style="color: #22d3ee;">üöÄ Capacity:</strong> Enables solving problems too large for single computers (storage/processing capacity)
  </p>
</div>

<div style="background: linear-gradient(135deg, #1e293b 0%, #334155 100%); padding: 30px; border-radius: 12px; margin: 25px 0; border-left: 6px solid #06b6d4; box-shadow: 0 4px 15px rgba(0,0,0,0.2);">
  <h3 style="color: #06b6d4; margin-top: 0; font-size: 1.8em; display: flex; align-items: center; gap: 10px;">
    <span style="background: #06b6d4; color: #0f172a; padding: 5px 15px; border-radius: 8px; font-weight: bold;">3</span>
    Module 3: Key Requirements & Limitations 
    <span style="color: #94a3b8; font-weight: normal; font-size: 0.75em; margin-left: auto;">(Arnav)</span>
  </h3>
  
  <p style="color: #e2e8f0; line-height: 1.8; font-size: 1.1em; margin: 15px 0;">
    <strong style="color: #06b6d4;">‚úÖ Independence Required:</strong> Tasks must be independent to run in parallel (no dependencies between steps)
  </p>
  <p style="color: #e2e8f0; line-height: 1.8; font-size: 1.1em; margin: 15px 0;">
    <strong style="color: #06b6d4;">‚è±Ô∏è Sequential Bottleneck:</strong> Sequential portions must still run in order (e.g., setup, initialization, combining results)
  </p>
  <p style="color: #e2e8f0; line-height: 1.8; font-size: 1.1em; margin: 15px 0;">
    <strong style="color: #06b6d4;">üìâ Diminishing Returns:</strong> Sequential portions limit maximum speedup (diminishing returns with more processors)
  </p>
  <p style="color: #e2e8f0; line-height: 1.8; font-size: 1.1em; margin: 15px 0;">
    <strong style="color: #06b6d4;">üîÑ System Overhead:</strong> Overhead from communication, coordination, and synchronization
  </p>
</div>

<div style="background: linear-gradient(135deg, #1e293b 0%, #334155 100%); padding: 30px; border-radius: 12px; margin: 25px 0; border-left: 6px solid #0ea5e9; box-shadow: 0 4px 15px rgba(0,0,0,0.2);">
  <h3 style="color: #0ea5e9; margin-top: 0; font-size: 1.8em; display: flex; align-items: center; gap: 10px;">

    Module 4: Execution Time Calculations 
    <span style="color: #94a3b8; font-weight: normal; font-size: 0.75em; margin-left: auto;">(Aaryav)</span>
  </h3>
  
  <p style="color: #e2e8f0; line-height: 1.8; font-size: 1.1em; margin: 15px 0;">
    <strong style="color: #0ea5e9;">üìä Sequential Formula:</strong> Sequential time = sum of all step times
  </p>
  <p style="color: #e2e8f0; line-height: 1.8; font-size: 1.1em; margin: 15px 0;">
    <strong style="color: #0ea5e9;">‚ö° Parallel Formula:</strong> Parallel time = sequential portion time + longest parallel task time
  </p>
  <p style="color: #e2e8f0; line-height: 1.8; font-size: 1.1em; margin: 15px 0; background: #0f172a; padding: 15px; border-radius: 8px; border-left: 4px solid #0ea5e9;">

  </p>

<div style="background: linear-gradient(135deg, #1e293b 0%, #334155 100%); padding: 30px; border-radius: 12px; margin: 25px 0; border-left: 6px solid #3b82f6; box-shadow: 0 4px 15px rgba(0,0,0,0.2);">
